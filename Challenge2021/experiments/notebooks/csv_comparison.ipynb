```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import os
```

```python
def clean_df(df):
  df = df.reset_index()
  df = df.drop(axis=0, index=0)
  df = df.drop(axis=1, columns=0)
  df = df.drop(axis=1, columns="index")
  df.index -=1
  return df
```

```python
def unique_elems(lst1, lst2):
    lst3 = [value for value in lst1 if not value in lst2]
    return lst3
```

```python
def cmp(df1, df2):
    rec_comb_union = []
    rec1_unique = []
    rec2_unique = []
    for i in tqdm(range(len(df1))):
      list1 = list(map(int, df1.loc[i][1].split()))
      list2 = list(map(int, df2.loc[i][1].split()))
      union_set = set(list1)
      union_set.update(list2)
      rec_comb_union.append(union_set)
      rec1_unique.append(unique_elems(list1, list2))
      rec2_unique.append(unique_elems(list2, list1))
    return rec_comb_union, rec1_unique, rec2_unique
```

```python
import csv
with open('/content/submission-48147.csv', 'r') as file:
    reader = csv.reader(file)
    df_48147 = clean_df(pd.DataFrame(reader))
with open('/content/submission-50908.csv', 'r') as file:
    reader = csv.reader(file)
    df_50908 = clean_df(pd.DataFrame(reader))
with open('/content/submission-50910.csv', 'r') as file:
    reader = csv.reader(file)
    df_50910 = clean_df(pd.DataFrame(reader))
```

```python
comparison = []
comparison.append(cmp(df_48147, df_50908))
comparison.append(cmp(df_48147, df_50910))
comparison.append(cmp(df_50908, df_50910))
```

```python
comparison_48147_50908_df = pd.DataFrame(list(zip(comparison[0][0],comparison[0][1],comparison[0][2])), columns=["Union", "Unique_48147", "Unique_50908"], )
comparison_48147_50908_df
```
